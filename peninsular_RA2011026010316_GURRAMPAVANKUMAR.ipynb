```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("books.csv")
df
```

```python
#Q1 NUMBER OF ROWS AND COLOUMNS
rows,columns=df.shape
print(" no of rows = ",rows)
print(" no of columns = ",columns)
```

```python
#Q2 How many books do not have an original title?
newdf=df['original_title'].isnull().sum()
print(f" books without original title={newdf}")
```

```python
#Q3 dropping null data from original_title 
filtered_data = df.dropna(subset=['original_title'])
print(filtered_data.info())
```

```python
#Q3 : How many unique books are present in the dataset ? Evaluate based on the 'book_id' after removing null values in the original_title column
unique_books_count = filtered_data['book_id'].nunique()
print("no of  unique books are present in the dataset : ",unique_books_count)
```

```python
#Q4  What is the average rating of all the books in the dataset based on ‘average_rating’? WITH ORIGINAL DATA 
average_rating=df['average_rating'].mean()
print(" Average rating of all books with original data=",{average_rating})
average_rating1 = filtered_data['average_rating'].mean()
print(" Average rating of all books with filtered data ",average_rating1)
```

```python
#Q5 Find the number of books published in the year ‘2000’ based on the ‘original_publication_year
numberofbooks_publish_2000=df[df['original_publication_year']==2000]
print(" number of books published in the year 2000 are : ",numberofbooks_publish_2000.shape[0])
numberofbooks_publish_2000_1=filtered_data[filtered_data['original_publication_year']==2000]
print(" number of books published in the year 2000 are : ",numberofbooks_publish_2000_1.shape[0])
```

```python
#Q6 Which book (title) has the maximum number of ratings based on ‘work_ratings_count
maximum_rating_count=filtered_data.loc[filtered_data['work_ratings_count'].idxmax()]['title']
print("maximum number of ratings :",maximum_rating_count)
```

```python
bins = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
df['rating_bucket']=pd.cut(df['average_rating'], bins, right=False, labels=False)
rating_counts = df['rating_bucket'].value_counts().sort_index()
plt.bar(rating_counts.index, rating_counts.values, width=0.3)
plt.xlabel('Rating Buckets')
plt.ylabel('Number of Books')
plt.xticks(range(len(bins)-1),[f'{x:.1f}-{x+0.5:.1f}' for x in bins[:-1]])
plt.title('Number of Books in Each Rating Bucket')
plt.show()
```
